```python
import numpy as np
import numpy as np
import matplotlib

from matplotlib import pyplot as plt

ts = pd.DataFrame(np.random.randn(1000,4), index=pd.date_range('1/1/2000',periods = 1000))
ts = ts.cumsum()
df = pd.DataFrame(np.random.randn(1000,4), index =ts.index, columns=['A','B','C','D'])
df = df.cumsum()
df.plot() 
plt.legend(loc = 'best')
```
